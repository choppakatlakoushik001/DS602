## Question 2: Wide data isn't usually helpful.
Convert this to a long format with the following columns:  
1. country  
2. year  
3. diagnosis method: mapping should be: 

```
  {'newsp':'Pulmonary Smear (+)'
  ,'newsn':'Pulmonary Smear (-)'
  ,'newep':'Extrapulmonary'
  ,'newrel':'Relapse'}   
 ```

4. gender: male or female.   
5. age: lower_age - higher_age, e.g., 0-14  
6. number of cases  

__Note__: You can use `replace` (https://pandas.pydata.org/docs/reference/api/pandas.DataFrame.replace.html)

````python
df
````

````python
# Code
#melting by country and year
df1=pd.melt(df, id_vars =['country','year'])

#creating gender column using apply function
df1['gender'] = df1['variable'].apply(lambda x: x.split('_')[-1][0])

#creating age column using apply function
df1['age'] = df1['variable'].apply(lambda x: x[-2:])

#making a dictionary for replacing
diagnosis_mapping = {
    'newsp': 'Pulmonary Smear (+)',
    'newsn': 'Pulmonary Smear (-)',
    'newep': 'Extrapulmonary',
    'newrel': 'Relapse'
}

#replacing '_' with empty space in variable column
df1['variable'] = df1['variable'].str.replace('_','')

#splitting the elements from variable column and assigning the values in a new column
df1['diagnosis_method'], remaining = df1['variable'].str.split('m|f').str

#mapping those values using the dictionary
df1['diagnosis_method'] = df1['diagnosis_method'].map(diagnosis_mapping)

#number of cases are values   
df1['number_of_cases'] = df1['value']

#removing variable and value columns
df2=df1.drop(['variable', 'value'], axis=1)
df2
````

# Question 3: Create a graph that shows when countries started to report TB cases.

````python
df2
````

````python
#replacing 0 with NaN in 'number_of_cases' column
df2['number_of_cases']=df2['number_of_cases'].replace(0,np.nan)

#dropping null values in 'number_of_cases' column
df3=df2.dropna(subset=['number_of_cases'])

#grouping the data by country and getting the first row of every group 
df4=df3.groupby('country',as_index=False).nth(0)

#drawing scatter plot between country and year
df4.plot(x='country', y='year', kind='scatter',rot=90)

# Adding labels and title
plt.xlabel('country')
plt.ylabel('year')
plt.title('starting year of TB cases for every country')

# Show the plot
plt.show()
````

````python
#plotting seperate graphs by dividing the dataframe into 10 smaller dataframes

#number of rows for every group
n=len(df4)//10

#create a list to store the smaller DataFrames
smaller_dataframes = []

#split the original DataFrame into 10 smaller DataFrames
for i in range(10):
    start_index = i*n
    end_index = (i+1)*n
    smaller_df = df4.iloc[start_index:end_index]
    smaller_dataframes.append(smaller_df)

for i in range(10):
    sd = smaller_dataframes[i].plot(x='country', y='year', kind='scatter',rot=90)
    sd.set_ylim(1980, 2013)

    # Add labels and title
    plt.xlabel('country')
    plt.ylabel('year')
    plt.title('starting year of TB cases for every country')

    # Show the plot
    plt.show()
````
